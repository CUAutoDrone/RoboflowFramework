```python
import os
import filetype

# train directory
train_dir = ''
# test directory
test_dir = ''
# validation directory
val_dir = ''

# Get images
train_images = [file_name for file_name in os.listdir(train_dir) if filetype.is_image(train_dir+file_name)]
test_images = [file_name for file_name in os.listdir(test_dir) if filetype.is_image(test_dir+file_name)]
val_images = [file_name for file_name in os.listdir(val_dir) if filetype.is_image(val_dir+file_name)]
```

```python
from roboflow import Roboflow

# Initialize the Roboflow object with your API key
rf = Roboflow(api_key="YOUR_PRIVATE_API_KEY")

# Retrieve your current workspace and project name
print(rf.workspace())
```

```python
# Specify the project for upload
project = rf.workspace("WORKSPACE_NAME").project("YOUR_PROJECT_ID")

# Upload the image to your project
project.upload("UPLOAD_IMAGE.jpg")

# Upload images to different splits
for image in train_images:
  project.upload(
      image=image,
      split="train",
      num_retry_uploads=3
  )

for image in test_images:
  project.upload(
      image=image,
      split="test",
      num_retry_uploads=3
  )

for image in val_images:
  project.upload(
      image=image,
      split="validation",
      num_retry_uploads=3
  )
```
